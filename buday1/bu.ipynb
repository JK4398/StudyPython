```python
from PIL import Image 
# 컬러이미지는 [R,G,B]를 갖고 있고 각 R,G,B는 0~255값임. 255.0.0은 빨강
fileName = folderName +'/'+ file_list[0]
img = Image.open(fileName)
img = img.resize((100,200))  #너비100, 높이 200/ 행은 200,열은 100
# np.shape(img) # [[1,2,3,[...,...   ],[   ]] 이런구조가 있다
# imgArray = np.array(img)
# #imgArray[:3] # = imgArray[:20]*0
# #imgArray 
# plt.imshow(imgArray[:20])

import matplotlib.pyplot as plt # 차트
imgArray = np.array(img)
re = imgArray*[2]  #-[100] [100] 할 시 사진 상 색상이 변경됨
plt.imshow(re)

val = [[1,0,0],[0,1,1],[0,0,1]]
for i in val:
     re = imgArray*i
     plt.imshow(re)
     plt.show() #위의 조건만 작성시 하나만 도출되지만, 여러 사진 도출을 원할 시 작성
```
